```python
import urllib.request
import zipfile
import io

import pandas as pd

g_cols = ['Glacier ID','Political Unit','Continent','Basin Code','Location Code','Glacier Code','Glacier Name','Latitude','Longitude','Primary Class','Glacier Source','Basin Count','Glacier Form','Glacier Activity','Activity Start','Activity End','Minimum Elevation','Minimum Elevation Exposed','Mean Elevation','Mean Elevation Accumulation','Mean Elevation Ablation','Maximum Elevation','Snow Line Elevation','Snow Line Accuracy','Glacier Area','Area Accuracy','Area Exposed','Mean Width','Mean Length','Maximum Length','Maximum Length Exposed','Maximum Length Ablation','Mean Depth','Depth Accuracy','Accumulation Orientation','Ablation Orientation','Topographic Map Year','Topographic Map Scale','Photograph Year']
glaciers = pd.read_csv('/content/drive/MyDrive/data/glacier/database.csv', sep=',', names=g_cols, usecols=g_cols)

glaciers.head(100)
```

# Nächster Gletscher am Äquator

```python
glaciers['Latitude'] = pd.to_numeric(glaciers['Latitude'], errors='coerce')

closest_glacier = glaciers.iloc[(glaciers['Latitude'] - 0).abs().idxmin()]

print(closest_glacier)
```

# Hochste Gletscher der Kontinente



## Europa



```python
glaciers['Maximum Elevation'] = pd.to_numeric(glaciers['Maximum Elevation'], errors='coerce')

european_glaciers = glaciers[glaciers['Continent'].str.strip().str.lower() == 'europe']
if not european_glaciers.empty:
    highest_glacier_europe = european_glaciers.loc[european_glaciers['Maximum Elevation'].idxmax()]
    print(highest_glacier_europe)
else:
    print("Keine Gletscher in Europa gefunden.")
```
